```python
import pandas as pd
import numpy as np
data=pd.read_csv('netflix_data.csv')
arr=data['cast'].str.split(", ")
arr.fillna(0,inplace=True)

def num_of_cast(x):
    
    if x!=0:
        return(len(x))
    else:
        return(0)
    
data["num_of_cast"]=arr.apply(num_of_cast)
data.head(15)
```
